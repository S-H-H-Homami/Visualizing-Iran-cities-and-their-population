```python
from mpl_toolkits.basemap import Basemap
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
file=pd.read_csv("D:\\Python\\DA-ML\\Session 8\\ir2.csv")
file.head()
```

```python
file.drop(columns=["country", "iso2", "admin_name", "population_proper","capital"],inplace=True)
file.dropna(inplace=True)
```

```python
m = Basemap(llcrnrlon=43.,llcrnrlat=24.,urcrnrlon=64.,urcrnrlat=41.,\
            resolution='i',projection='merc')
# m.bluemarble()
m.etopo()
# m.shadedrelief()


pop=np.sum(file["population"])
for i in range(0,len(file)):
        m.scatter(file["lng"][i],file["lat"][i],s=file["population"][i]*4000/pop,latlon=True)
        xpt,ypt = m(file["lng"][i],file["lat"][i])
        plt.text(xpt,ypt,file["city"][i],fontsize=6)
```
